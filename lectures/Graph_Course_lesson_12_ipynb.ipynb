## Определения

**Гамильтоновым путём** (англ. Hamiltonian path) называется простой путь, проходящий через каждую вершину графа ровно один раз.

**Гамильтоновым циклом** (англ. Hamiltonian cycle) называют замкнутый гамильтонов путь.

Граф называется **полугамильтоновым** (англ. Semihamiltonian graph), если он содержит гамильтонов путь.

Граф называется **гамильтоновым** (англ. Hamiltonian graph), если он содержит гамильтонов цикл.

$$d_k \le k < n/2 ⇒ d_{n-k} \ge n-k $$

По принципу Дирихле всегда найдутся две смежные вершины $t_1,t_2$

## Задача Коммивояжера

**Задача коммивояжёра** (или TSP от англ. travelling salesman problem) — одна из самых известных задач комбинаторной оптимизации, заключающаяся в поиске самого выгодного маршрута, проходящего через указанные города хотя бы по одному разу с последующим возвратом в исходный город. В условиях задачи указываются критерий выгодности маршрута (кратчайший, самый дешёвый, совокупный критерий и тому подобное) и соответствующие матрицы расстояний, стоимости и тому подобного. Как правило, указывается, что маршрут должен проходить через каждый город только один раз — в таком случае выбор осуществляется среди гамильтоновых циклов.

## Метод ветвей и границ

https://habr.com/ru/articles/560468/

Относится к классу методлов "разделяй и властвуй"

Все перебираемые варианты разбиваются на классы.

Для каждого класса находится оценка снизу всех решений этого класса

Если оценка превышает стоимость ранее найденного решения, все варианты классса отбрасываются




Пусть  $х_{ij} = 1$, если путешественник переезжает из $i$-ого города в $j$-ый и $х_{ij} =0$, если $j$-ый город не посещается. Условно введем $(n+1)$-й город, совмещенный с 1-м городом, т.е. расстояния от $(n+1)$-го города до любого другого, отличного от первого, равны расстояниям от первого города. При этом, если из первого города можно лишь выйти, то в $(n+1)$-й  город можно лишь прийти. Введем дополнительные целые переменные, равные номеру посещения этого города на пути. $u_1=0$, $u_{n+1}=n$. Для того, чтобы избежать замкнутых путей, выйти из первого города и вернуться в $(n+1)$ введем дополнительные ограничения, связывающие переменные $x_{ij}$ и переменные $u_i$ ($u_i$ целые неотрицательные числа)

$$min \sum^n_{i=1}\sum^n_{j=1}c_{ij}x_{ij} $$
где
$$\sum_{i=1}^{n}x_{ij} = 1, \quad 1\le j \le n$$
$$\sum_{j=1}^{n}x_{ij} = 1, \quad 1\le i \le n$$
$$\sum^{}_{i \in S}\sum_{j \ne S}^{n}x_{ij} \ge 1, \quad S \ne \emptyset, S \subset \{1,...,n\}$$
$$x_{ij} \in \{0,1\}, \quad 1\le i,j \le n$$

На каждом шаге алгоритма решения задачи коммивояжера выполняются (определяются):
* Формирование множества $S$ клеток с нулевыми значениями в матрице и их характеристик $d_i$, $d_j$, $∑_{ij}$ в форме плоской таблицы;

* Вычисляются max $∑_{ij}$ и определяется клетка $(k, l)$ с $max$ которая не включается в маршрут;

* Модифицируется матрица: из матрицы удаляются $k$-я строка и $l$-й столбец клетки $(k, l)$;

* Вычисляются оценки НГЦФ (Нижней границы Целевой функции)  для позитивного и негативного узлов дерева поиска решения;

* В маршрут включается элемент $(k, l)$ с предпочтительной (меньшей) оценкой $ω(k, l)$ НГЦФ

* Сравниваются оценки между собой исходная $Q^*$ и вычисленная $Н$ на текущем шаге;

* Вычисляется размерность $dimС_{[n]}$ матрицы и проверяется равенство этой размерности с $dim 2×2$;
* Формируется множество В висячих узлов дерева решений с их оценками НГЦФ.

* Приводится фрагмент дерева поиска решений для очередного шага.

**Итерационные  шаги**: Формирование тура $Т$ начинается выбором 1-й дуги  $(k, l)$ графа дорог  для  включения  её в  маршрут  T, оформляемый как последовательность дуг $Т = \{(a,b)(c,d)(ef)(hg).....\}$.

##Эвристические методы поиска Гамильтонова цикла

### Алгоритм ближайшего соседа

На каждомм шаге жадно добавляем к последнем ребру цепи любое смежное ребро, пока не полуим цикл

Как правило результат далек от идеального )

 $∀ k>0 ∃$ граф в ккотором вес гамилтонова цикла построенного по этому алгоритму будет в k раз больше минимального

### Алгоритм ближайшей вставки


Возьмум любой цикл (не обязательно гамильтоновый) в графе

Будем увеличивать его включая в него ближайшие вершины

Вес построенного цикла не более чем в 2 раза больше веса минимального гамилтонового цикла


### Алгоритм локальной оптимизации (похож на имитацию отжига)

В имеющемся гамильтоновом цикле для всех вершин $i,j: 1 \le i < p $ делается попытка поменять местами вершины $i$ и $j$. Если перестановка уменьшает вес цикла то улучшаем решение

### Алгоритм Эйлера

работает при условии

$$∀i,j:W(i,j) + W(j,k) \le W(i,k)$$

1. Cтроится остов (Крускал или Прим)
2. Путем дублирования каждого ребра остов превращается в эйлеров граф
3. Находим Эйлеров цикл с помощью алгоритма Флери
4. Преобразуем эйлеров цикл в гамильтонов: Сокращаем количество вершин в таким образом, чтобы в итоговом цикле каждая вершина встречалась только один раз

Вес решения не превосходит удвоенного веса оптимального решения


### Алгоритм Кристофера


1. Cтроится остов (Крускал или Прим)
2. На множестве вершин остова нечетной степени находится паросочетание минимального веса
3. Остов преобразуется в Эйлеров граф путем присоединения ребер соответствующих найденному паросочетанию
4. Преобразуем эйлеров цикл в гамильтонов: Сокращаем количество вершин в таким образом, чтобы в итоговом цикле каждая вершина встречалась только один раз
