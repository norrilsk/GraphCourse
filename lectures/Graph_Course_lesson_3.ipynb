## Бинарное дерево поиска

Такое дерево может быть представлено при помощи связанной структуры данных, в которой каждый узел является объектом. В дополнение к полям ключа $key$ и сопутствующих данных, каждый узел содержит поля $left$, $right$ и $p$, которые указывают на левый и правый дочерние
узлы и на родительский узел соответственно.

Если $x$ — узел бинарного дерева поиска, а узел y находится в левом поддереве $x$, то $key [y] \leq key [x]$. Если узел y находится в правом
поддереве x, то $key [x] \leq key [y]$.

Получить все ключи в отсортированном порядке можно применяя $inorder$ нумерацию


image.png

### Поиск
$TREE\_SEARCH(x, k)$<br>
$\quad$ $\textbf{if} \;  x = NIL  \; \textbf{or} \; k = key[x]$<br>
$\quad \quad  \textbf{then}\; \textbf{return}\; x$<br>
$\quad \textbf{if} \; k < key[x]$<br>
$\quad \quad \textbf{then} \;\textbf{return} \;TREE\_SEARCH(left[x], k)$<br>
$\quad \quad \textbf{else} \; \textbf{return} \; TREE\_SEARCH(right[x], k)$

<details>
<summary>Сложность Поиска элемента:</summary>
&emsp;
$O(h)$, где $h$ - высота дерева
</details>

### Поиск минимума и максимума
$TREE\_MINIMUM(x)$ <br>
$\quad \textbf{while} \; left[x] \neq NIL$ <br>
$\quad \quad \textbf{do} \;x ← left[x]$ <br>
$\quad \textbf{return} \; x$

$TREE\_MAXIMUM(x)$ <br>
$\quad \textbf{while} \; right[x] \neq NIL$ <br>
$\quad \quad \textbf{do} \;x ← right[x]$ <br>
$\quad \textbf{return} \; x$

<details>
<summary>Сложность Поиска минимального элемента:</summary>
&emsp;
$O(h)$, где $h$ - высота дерева
</details>

### Поиск следующего

$TREE\_SUCCESSOR(x)$<br>
$\quad \textbf{if} \;right[x] \neq NIL$<br>
$ \quad \quad  \textbf{then return}\; TREE\_MINIMUM(right[x])$ <br>
$ \quad y ← p[x]$ <br>
$\quad \textbf{while}\; y \neq NIL\; и\; x == right[y]$ <br>
$\quad \quad\quad \textbf{do}\; x ← y$ <br>
$\quad\quad\quad \quad y ← p[y]$ <br>
$ \quad \textbf{return}\; y$

<details>
<summary>Сложность Поиска следующего элемента:</summary>
&emsp;
$O(h)$, где $h$ - высота дерева
</details>

## Красно-черные деревья

**красно-черное дерево** - бинарное дерево поиска которое удовлетворет следующим 5 свойствам:
1. Каждый узел является красным или черным.
2. Корень дерева является черным.
3. Каждый лист дерева (NIL) является черным.
4. Если узел — красный, то оба его дочерних узла — черные.
5. Для каждого узла все пути от него до листьев, являющихся потомками
данного узла, содержат одно и то же количество черных узлов.



image.png

**Лемма 13.1**. Красно-черное дерево с n внутренними узлами имеет высоту не
более чем 2 lg (n + 1).

<details>
<summary>Доказательство</summary>
&emsp;


 Начнем с того, что покажем, что поддерево любого узла x содержит как минимум $2^{bh(x)} − 1$ внутренних узлов. Докажем это по индукции по
высоте x. Если высота x равна 0, то узел x должен быть листом (nil [T]) и п оддерево узла x содержит не менее $2^{bh(x)} − 1 = 2^0 − 1 = 0$ внутренних узлов.


Теперь рассмотрим узел x, который имеет положительную высоту и представляет собой внутренний узел с двумя потомками. Каждый потомок имеет черную
высоту либо bh (x), либо bh (x) − 1, в зависимости от того, является ли его цвет
соответственно красным или черным. Поскольку высота потомка x меньше, чем высота самого узла x, мы можем использовать предположение индукции и сделать
вывод о том, что каждый из потомков x имеет как минимум $2^{bh(x)} − 1$ внутренних узлов. Таким  образом, дерево с корнем в вершине x содержит как минимум
$(2^{bh(x)−1} − 1) + (2^{bh(x)−1} − 1) + 1 = 2^{bh(x)}−1$ внутренних узлов, что и доказывает
наше утверждение.


Для того чтобы завершить доказательство леммы, обозначим высоту дерева
через h. Согласно свойству 4, по крайней мере половина узлов на пути от корня
к листу, не считая сам корень, должны быть черными. Следовательно, черная
высота корня должна составлять как минимум h/2; следовательно,

$$ n \geq 2^{h/2} -1$$

Перенося 1 в левую часть и логарифмируя, получим, что $lg (n + 1) \geq h/2,\;или\; h \leq 2 lg (n + 1)$.

### Повороты


image.png

$LEFT\_ROTATE(T, x)$<br>
$\quad y ← right[x]  \quad // \quad Устанавливаем \;y$<br>
$\quad right[x] ← left[y] \quad // \quad Левое \; поддерево \;y \;становится \; правым \; поддеревом \;x$<br>
$\quad \textbf{if} \; left[y] \neq nil [T]$<br>
$\quad \quad \textbf{then} \;  p[left[y]] ← x$<br>
$\quad p[y] ← p[x] \quad // \quad  Перенос \; родителя \; x \; в \;  y$<br>
$\quad \textbf{if}\; p[x] = nil [T]$<br>
$\quad \quad \textbf{then}\; root[T] ← y$<br>
$\quad \quad \textbf{else} \; \textbf{if} \; x = left[p[x]]$<br>
$\quad \quad \quad \textbf{then} \;left[p[x]] ← y$<br>
$\quad \quad \quad \textbf{else} \;right[p[x]] ← y$<br>
$\quad left[y] ← x \quad // \quad  x — левый \;дочерний \; y$<br>
$\quad p[x] ← y$<br>

### Вставка

* Сложность вставки $O(lg(N))$
* Цвет нового элемента всегда **красный**
* В конце процедуры вызываем RB_INSERT_FIXUP, которая перекрашивает узлы и выполняет повороты.




$RB\_INSERT(T, z)$<br>
$\quad y ← nil [T]$<br>
$\quad x ← root[T]$<br>
$\quad \textbf{while}\; x \neq nil [T]$<br>
$\quad \quad \textbf{do} y ← x$<br>
$\quad \quad \quad \textbf{if} \;key[z] < key[x]$<br>
$\quad \quad \quad \quad\textbf{then}\; x ← left[x]$<br>
$\quad \quad \quad \quad \textbf{else} \; x ← right[x]$<br>
$\quad p[z] ← y$<br>
$\quad \textbf{if} \; y = nil [T]$<br>
$\quad \quad \textbf{then}\; root[T] ← z$<br>
$\quad \quad \textbf{else}\; \textbf{if}\;  key[z] < key[y]$<br>
$\quad \quad \quad \textbf{then}\; left[y] ← z$<br>
$\quad \quad \quad \textbf{else}\; right[y] ← z$<br>
$\quad left[z] ← nil [T]$<br>
$\quad right[z] ← nil [T]$<br>
$\quad color [z] ← RED$<br>
$\quad RB\_INSERT\_FIXUP(T, z)$<br>

$RB\_INSERT\_FIXUP(T, z)$<br>
$\quad \textbf{while}\; color [p[z]] = RED$<br>
$\quad \quad \textbf{do}\; \textbf{if}\; p[z] = left[p[p[z]]]$<br>
$\quad \quad\quad \textbf{then}\; y ← right[p[p[z]]]$<br>
$\quad \quad\quad \quad \textbf{if}\; color [y] = RED$<br>
$\quad \quad\quad \quad \quad \textbf{then}\; color [p[z]] ← BLACK  \quad $ //  Случай 1<br>
$\quad \quad\quad \quad \quad \quad color [y] ← BLACK  \quad $ //  Случай 1<br>
$\quad \quad\quad \quad \quad \quad color [p[p[z]]] ← RED  \quad $ //  Случай 1<br>
$\quad \quad\quad \quad \quad \quad z ← p[p[z]]  \quad $ //  Случай 1<br>
$\quad \quad\quad \quad \quad \textbf{else}\; \textbf{if}\; z = right[p[z]]$<br>
$\quad \quad\quad \quad \quad \quad \textbf{then}\; z ← p[z]  \quad $ //  Случай 2<br>
$\quad \quad\quad \quad \quad \quad \quad LEFT\_ROTATE(T, z) \quad $ //  Случай 2<br>
$\quad \quad\quad \quad \quad \quad  color [p[z]] ← BLACK \quad $ //  Случай 3<br>
$\quad \quad\quad \quad \quad \quad  color [p[p[z]]] ← RED \quad $ //  Случай 3<br>
$\quad \quad\quad \quad \quad \quad  RIGHT\_ROTATE(T, p[p[z]]) \quad $ //  Случай 3<br>
$\quad \quad\quad \textbf{else}\;$  // (то же, что и в “then”, с заменой left на right и наоборот) <br>
$\quad color [root[T]] ← BLACK$<br>

### Удаление

$RB\_DELETE(T, z)$<br>
$\quad \textbf{if} \; left[z] = nil [T] \; или\;  right[z] = nil [T]$<br>
$\quad \quad \textbf{then} \; y ← z$<br>
$\quad \quad \textbf{else} \; y ← TREE\_SUCCESSOR(z)$<br>
$\quad \textbf{if} \; left[y] \neq nil [T]$<br>
$\quad \quad \textbf{then} \; x ← left[y]$<br>
$\quad \quad \textbf{else} \; x ← right[y]$<br>
$\quad p[x] ← p[y]$<br>
$\quad \textbf{if} \; p[y] = nil [T]$<br>
$\quad \quad \textbf{then} \; root[T] ← x$<br>
$\quad \quad \textbf{else} \; \textbf{if}\; y = left[p[y]]$<br>
$\quad \quad \quad \textbf{then} \; left[p[y]] ← x$<br>
$\quad \quad \quad \textbf{else} \; right[p[y]] ← x$<br>
$\quad \textbf{if} \; y \neq z$<br>
$\quad \quad \textbf{then} \; key[z] ← key[y]$<br>
$\quad \quad \quad $ Копируем сопутствующие данные y в z<br>
$\quad \textbf{if} \; color [y] = BLACK$<br>
$\quad \quad \textbf{then} \; RB\_DELETE\_FIXUP(T, x)$<br>
$\quad \textbf{return} \; y$<br>

$RB\_DELETE\_FIXUP(T, x)$<br>
$\quad \textbf{while}\; x \neq root[T] \; и\;  color [x] = BLACK$<br>
$\quad \quad \textbf{do}\; \textbf{if}\; x = left[p[x]]$<br>
$\quad \quad \quad \textbf{then}\; w ← right[p[x]]$<br>
$\quad \quad \quad \quad \textbf{if}\; color [w] = RED$<br>
$\quad \quad \quad \quad \quad \textbf{then}\; color [w] ← BLACK \quad $ // Случай 1<br>
$\quad \quad \quad \quad \quad \quad color [p[x]] ← RED \quad $ // Случай 1<br>
$\quad \quad \quad \quad \quad \quad LEFT\_ROTATE(T, p[x]) \quad $ // Случай 1<br>
$\quad \quad \quad \quad \quad \quad w ← right[p[x]] \quad $ // Случай 1<br>
$\quad \quad \quad \quad \textbf{if}\; color [left[w]] = BLACK \; и \; color [right[w]] = BLACK$<br>
$\quad \quad \quad \quad \quad \textbf{then}\; color [w] ← RED \quad $ // Случай 2<br>
$\quad \quad \quad \quad \quad  \quad x ← p[x] \quad $ // Случай 2<br>
$\quad \quad \quad \quad \quad \textbf{else}\; \textbf{if}\; color [right[w]] = BLACK$<br>
$\quad \quad \quad \quad \quad \quad \textbf{then}\; color [left[w]] ← BLACK \quad $ // Случай 3<br>
$\quad \quad \quad \quad \quad \quad \quad color [w] ← RED \quad $ // Случай 3<br>
$\quad \quad \quad \quad \quad \quad \quad RIGHT\_ROTATE(T,w) \quad $ // Случай 3<br>
$\quad \quad \quad \quad \quad \quad \quad w ← right[p[x]] \quad $ // Случай 3<br>
$\quad \quad \quad \quad \quad \quad color [w] ← color [p[x]] \quad $ // Случай 4<br>
$\quad \quad \quad \quad \quad \quad color [p[x]] ← BLACK \quad $ // Случай 4<br>
$\quad \quad \quad \quad \quad \quad color [right[w]] ← BLACK \quad $ // Случай 4<br>
$\quad \quad \quad \quad \quad \quad LEFT\_ROTATE(T, p[x]) \quad $ // Случай 4<br>
$\quad \quad \quad \quad \quad \quad x ← root[T] \quad $ // Случай 4<br>
$\quad \quad \quad \textbf{else}$ (то же, что и в “then”, с заменой left на right и наоборот)<br>
$\quad color [x] ← BLACK$<br>
